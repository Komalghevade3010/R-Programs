```python

import pandas as pd
import matplotlib.pyplot as plt
#from sklearn import datasets
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.metrics import classification_report
from sklearn import preprocessing
```

```python
mobile = pd.read_csv('/content/Mobile Data.csv')
```

```python
mobile.head()
```

```python
x=mobile.iloc[:,0:7]
y=mobile['price_range']
```

```python
x
```

```python
y
```

```python
colnames=list(mobile.columns)
colnames
```

```python
x_train, x_test,y_train,y_test = train_test_split(x,y, test_size=0.2,random_state=0)
```

```python
x_train
```

```python
x_train.shape
```

```python
y_train
```

```python
y_test
```

```python
y_test.shape
```

```python
model = DecisionTreeClassifier(criterion = 'entropy',max_depth=2,random_state=0) # C5.0, max_depth of tree = 3.
# To ensure that you get consistent results every time you run your Decision Tree model,
# you can set the random_state=0 parameter for the Decision Tree itself.
model.fit(x_train,y_train)
```

```python
tree.plot_tree(model)
```

```python
y_train.value_counts()
```

```python
fn=['battery_power',
 'four_g',
 'int_memory',
 'ram',
 'talk_time',
 'touch_screen',
 'wifi',
 'price_range']
cn=['0','1','2','3']
 # Class names
#fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (3,3), dpi=300)
tree.plot_tree(model,
               feature_names = fn,
               class_names=cn,
               filled = True)
```

```python
#Predicting on test data
preds = model.predict(x_test) # predicting on test data set
pd.Series(preds).value_counts() # getting the count of each category
```

```python
model.feature_importances_
```

```python
pd.crosstab(y_test,preds)
```
