```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
mobile=pd.read_csv("/content/Mobile Data.csv")
mobile.head()
```

```python
mobile.shape
```

```python
mobile.drop(["clock_speed","blue","n_cores","talk_time"],inplace=True,axis=1)
```

```python
mobile.shape
```

```python
mobile.info()
```

```python
mobile=mobile.dropna()
mobile.shape
```

```python
mobile.head()
```

```python
x=mobile.iloc[:,0:16]
y=mobile.iloc[:,16]
```

```python
x
```

```python
y
```

```python
classifier=LogisticRegression()
classifier.fit(x,y)
```

```python
y_pred=classifier.predict(x)
```

```python
y_pred
```

```python
y_pred_df=pd.DataFrame({'y':y,'Yhat':classifier.predict(x)})
```

```python
y_pred_df
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix=confusion_matrix(y,y_pred)
print(confusion_matrix)
```

```python
from sklearn.metrics import roc_curve
from sklearn.metrics import roc_auc_score
```

```python
fpr ,tpr,thresholds = roc_curve(y,classifier.predict_proba(x)[:,3],pos_label=3)
```

```python
auc = roc_auc_score(y, classifier.predict_proba(x),multi_class='ovr')
```

```python
import matplotlib.pyplot as plt
plt.plot(fpr, tpr, color='red')
plt.plot([0, 1], [0, 1], 'k--')
plt.xlabel('False Positive Rate or [1 - True Negative Rate]')
plt.ylabel('True Positive Rate')
plt.show()
```
